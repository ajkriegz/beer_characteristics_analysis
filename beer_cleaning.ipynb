```python
# Import dependencies.
import pandas as pd
import os
```

```python
# For more information on our datasets, see:
# https://www.kaggle.com/datasets/ruthgn/beer-profile-and-ratings-data-set
# https://www.kaggle.com/datasets/rdoume/beerreviews
# https://www.kaggle.com/datasets/stephenpolozoff/top-beer-information
```

```python
# Import the beer dataset from csv.
file = os.path.join("Resources", "beer_profile_and_ratings.csv")

# Read the CSV.
df = pd.read_csv(file)

# Show the dataframe.
df
```

```python
df["Style"].unique()
```

```python
df[df["review_overall"] >= 4]
```

```python
# Get the column names.
df.columns
```

```python
# Clean up the column names.
df.rename(columns = {"Style":"beer_style", "Beer Name (Full)":"beer_name_full",
    "ABV":"abv", "Min IBU":"min_ibu", "Max IBU":"max_ibu", "Astringency":"astringency",
    "Body":"body", "Alcohol":"alcohol", "Sweet":"sweet", "Sour":"sour",
    "Salty":"salty", "Fruits":"fruits", "Hoppy":"hoppy", "Spices":"spices",
    "Malty":"malty"}, inplace=True)

# Drop unnecessary columns.
df = df.drop(["Name", "Brewery", "Description"], axis=1)

# Show the dataframe.
df
```
