```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
```

```python
# Import and read the data
file = os.path.join("Resources", "beer_profile_and_ratings.csv")
df = pd.read_csv(file)

file2 = os.path.join("Resources", "beer_styles.csv")
df2 = pd.read_csv(file2)
```

```python
df.rename(columns = {"Style":"Beer_Style", "Beer Name (Full)":"Beer_Name_Full", "Min IBU":"Min_IBU", "Max IBU":"Max_IBU"}, inplace=True)

df = df.drop(["Name", "Description"], axis=1)

df
```

```python
# Find out how many unique breweries there are
len(df["Brewery"].unique())
```

```python
# Viz 1 -- No. breweries compared to no. beers on list (stacked bar?)
# https://www.geeksforgeeks.org/create-a-stacked-bar-plot-in-matplotlib/

# unique brewery values
breweries = len(df["Brewery"].unique())
beers = len(df["Beer_Name_Full"].unique())

x = "Number of Beer Reviews (Orange) vs. Number of Breweries in Database (Off White)"
y1 = breweries
y2 = beers

# Add legend?

# plot stacked bars
plt.bar(x, y2, color="orange")
plt.bar(x, y1, bottom = y2, color = "cornsilk")
plt.show()
```

```python
# Viz 2 -- Reviews on multiple line graph
review_df = df[["review_aroma", "review_appearance", "review_palate", "review_taste", "review_overall"]]

review_df.plot(title ="Reviews", ylabel ="Number of Reviews",
                     color =["lightskyblue", "lightcoral","gold"], 
                     xlabel ="", 
                     figsize =(25,6))

plt.legend(loc ='right')

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# plt.savefig("analysis/beer_summary.png")

# Scatter plot of distribution? Ranges?
```

```python
# Viz 3 -- Review distribution
numbers = {"review_score" : [5,4,3,2,1], "number_reviews": [24,2537,563,71,2]}
numbers_df = pd.DataFrame(numbers)
numbers_df
```

```python
# https://www.geeksforgeeks.org/create-lollipop-charts-with-pandas-and-matplotlib/

# Create an empty chart
fig, axes = plt.subplots(figsize=(8, 6))

# Plot using plt.stem
axes.stem(numbers_df["review_score"], numbers_df["number_reviews"],
          use_line_collection=True, basefmt=' ')

axes.set_ylim(0)

# Details and formatting 
plt.title("Distribution of Reviews")
plt.ylabel('Number of Reviews')
plt.xlabel('Review (Rounded)')
plt.xticks(numbers_df["review_score"])

plt.show()
```

```python
rounded_reviews = df["review_overall"].round(0)
rounded_reviews = rounded_reviews.sort_values(ascending=False)

rounded_reviews.value_counts()
```

```python
# Viz 4 -- Style Distribution
# Get the count of beers of each style
number_styles = df2["Style"].value_counts()
number_styles = number_styles.sort_index(ascending=False)
number_styles
```

```python
styles = df2["Style"].sort_values(ascending=False)
styles = styles.unique()
styles
```

```python
# https://www.geeksforgeeks.org/create-lollipop-charts-with-pandas-and-matplotlib/
# Create an empty chart
fig, axes = plt.subplots(figsize=(8, 6))

# Plot
axes.hlines(styles, xmin=0, xmax=number_styles)
axes.plot(number_styles, styles, 'o')

axes.set_xlim(0)

# Details and formatting 
plt.title("Distribution of Beer Styles")
plt.xlabel('Number of Beers in List')
plt.ylabel('Beer Style')

plt.show()
```

```python
# Viz 5 -- Scatter Plot
review_df = df[["review_aroma", "review_appearance", "review_palate", "review_taste", "review_overall"]]

review_df.plot(title ="Reviews", ylabel ="Number of Reviews",
                     color =["lightskyblue", "lightcoral","gold"], 
                     xlabel ="", 
                     figsize =(25,6))

plt.legend(loc ='right')

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
```

```python
x_labels = ["Aroma", "Appearance",
    "Mouthfeel", "Taste", "Overall"] 
fig, ax = plt.subplots(figsize=(10,6))

ax.set_title("Review Distribution", fontsize=20)
ax.set_ylabel("Review Score", fontsize=14)
ax.set_xlabel("Review Criteria", fontsize=14)
ax.boxplot(review_df, labels=x_labels)
plt.show()
```
